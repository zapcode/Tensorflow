```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Stock.csv")
X_train = df.iloc[:,1].values
Y_train = df.iloc[:,-1].values
```

```python
samples = X_train.shape[0]
```

```python
X = tf.placeholder("float64")
Y = tf.placeholder("float64")
```

```python
Weight = tf.Variable(np.random.rand(1))
biases = tf.Variable(np.random.rand(1))
```

```python
pred  = tf.add(tf.multiply(Weight,X),biases)
```

```python
cost = (tf.pow(pred-Y,2))/2*samples
```

```python
optimizer = tf.train.AdamOptimizer(0.01).minimize(cost)
```

```python
init = tf.global_variables_initializer()
```

```python
w = 0;
b = 0;
with tf.Session() as sess:
    sess.run(init)
    for epochs in range(10000):
        sess.run(optimizer,feed_dict = {X:X_train,Y:Y_train}) 
        plt.plot(X_train,Y_train,color='b')
        plt.plot(X_train,X_train*sess.run(Weight)+sess.run(biases),color = 'r')
    print("Weight = {}, Biases = {}".format(sess.run(Weight),sess.run(biases)))
    w = sess.run(Weight)
    b = sess.run(biases)
```
